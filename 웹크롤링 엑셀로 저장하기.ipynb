```python
#종준님 코드
#네이버 한식 맛집 자동크롤링 코드
import time
import requests
import openpyxl
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
import xlsxwriter as xw

wb = openpyxl.Workbook()
url = 'https://www.naver.com/'
driver = webdriver.Chrome('c:\chromedriver.exe')
driver.get(url)
# driver.find_elements_by_css_selector("button.btn_close")[1].click()
# search_box = driver.find_element_by_css_selector("input#query")

names = []
rating = []
sheet = wb.active
sheet.append(["상호", "별점"])

##### 한식 메뉴 페이지 #####
search_box = driver.find_element_by_css_selector("input#query") # 검색창 마우스 클릭
search_box.send_keys('효령로 한식 맛집') # 원하는 검색어 입력
driver.find_element_by_xpath('//*[@id="search_btn"]').click()

print("-------------- 검색 중입니다. ---------------")

time.sleep(1)

##### 한식 메뉴 검색 #####
for i in range(7):
    stores = driver.find_elements_by_css_selector("div.list_item_inner")
    time.sleep(2)
    for store in stores:
        name = store.find_element_by_css_selector(".name span").text
        try:
            rating = store.find_element_by_css_selector("span.rating").text
        except:
            rating = "None"

        print(name)
        print(rating)
        sheet.append([name, rating])
    
    #다음페이지 넘기기
    time.sleep(2)
    search_button = driver.find_element_by_css_selector("#place_main_ct > div > section:nth-child(1) > div > div.place_list > div.list_area > div > a.spnew_bf.cmm_pg_next.on").click()
    time.sleep(1)
print("---------------- 검색이 끝났습니다. -----------------")

time.sleep(2)
# driver.quit() #페이지 종료
# wb.save("data2.xlsx") #엑셀 데이터 저장
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import pandas as pd
import time


driver = webdriver.Chrome("c:/chromedriver.exe")
driver.get("https://v4.map.naver.com")
driver.find_elements_by_css_selector("button.btn_close")[1].click()
search_box = driver.find_element_by_css_selector("input#search-input")

names = []
addrs = []
cates = []

list = ["효령로 한식 맛집","효령로 일식 맛집","효령로 분식 맛집","효령로 중식 맛집"]
for a in list:
    search_box.send_keys(a) 
    search_button = driver.find_element_by_css_selector("button.spm")
    search_button.click()
    print("**********"+ a +"**********")
    time.sleep(2)
    
    for n in range(2, 6):
        stores = driver.find_elements_by_css_selector("div.lsnx")
        
        for store in stores:
            name = store.find_element_by_css_selector("dt > a").text
            addr = store.find_element_by_css_selector("dd.addr").text
            cate = store.find_element_by_css_selector("dd.cate").text
            print(name)
            print(addr)
            print(cate)
            names.append(name)
            addrs.append(addr)
            cates.append(cate)

        page_bar = driver.find_elements_by_css_selector('div.paginate > *')
        try:
            if n%5 != 0:
                page_bar[n%5+1].click()
            else:
                page_bar[6].click()
        except:
            print("***************수집완료******************")
            break
            
        time.sleep(2)
        search_box.clear()
        
data = pd.DataFrame({"상호":names, "주소":addrs, "메뉴":cates})
data.to_excel("data1.xlsx")
```
